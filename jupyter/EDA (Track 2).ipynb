### Age

#### A
- 전부 기혼
- 중고생 자녀가 있을 확률이 제일 높음
- 여자는 전업주부일 확률이 높음

<br>

#### B
- 전부 기혼
- 유아 자녀는 아예 없음
- 초등학교 자녀를 둔 사람들의 확률이 높음
- 중고등학교 자녀를 둘 가능성도 높음
- 대학생 자녀도 확률이 낮지 않음
- 여자는 전업주부일 확률이 A보다는 낮음

<br>

#### C
- 기혼을 안 하기 시작 (low가 낮음)
- 기혼한지 얼마 안 된 어린 부부들
- 유아 자녀, 초등학교 자녀 스코어만 높음

<br>

#### D
- 기혼이 mid, high만 있음
- 초, 중고등학교 자녀를 둔 사람들

<br>

#### E
- 기혼 안 한 사람의 비중이 더 높음
- 유아자녀 스코어만 있음

<br>

#### F
- 전부 기혼
- 유아는 거의 없고
- 초중고대학교 자녀 모두 있음

<br>

#### G
- 전부 기혼
- 초중고 비중이 제일 높고 대학교 자녀도 있음

<br>

#### H
- 전부 기혼
- 자녀가 없거나 있어도 대학생 자녀들

```python
for age in sorted(trend_w_demo['연령대'].unique().tolist()):
    print("#"*50,age,"#"*50)
    df = trend_w_demo[trend_w_demo['연령대']==age]
    df_male = df[df['성별구분']==0]
    df_female = df[df['성별구분']==1]

    cols = ['기혼스코어', '유아자녀스코어', '초등학생자녀스코어', '중고생자녀스코어', '대학생자녀스코어', '전업주부스코어']
    for col in cols:
        print("@@@남자@@@")
        print(df_male[col].value_counts().sort_index(),'\n')
        print("@@@여자@@@")
        print(df_female[col].value_counts().sort_index())
        print("*"*60, '\n')
```
